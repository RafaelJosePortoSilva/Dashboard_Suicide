```python
from pyspark.sql import SparkSession
```

```python
spark = SparkSession.builder.getOrCreate()
df = spark.read.option('header',True) \
        .csv("CSV/suicide_rates_1990-2022.csv",inferSchema=True)
df.printSchema()
```

```python
df1 = df.groupBy('CountryName').sum('SuicideCount').sort('sum(SuicideCount)',ascending=False)
```

```python
df.select('CountryName').distinct().count()
```

```python
!pip install seaborn
```

```python
import seaborn as sns

barplot = df1.toPandas()

sns.barplot(data=barplot.head(20),
           x='CountryName',
           y='sum(SuicideCount)')
```
